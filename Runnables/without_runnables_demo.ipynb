```python
import random

class dupLLM:
    def __init__(self):
        print("LLM created")
    
    def predict(self,prompt):
        response_list=[
            "IPL is a cricket league",
            "FIFA is a football league",
            "Wimbledon is a tennis league"
        ]
        return {"response":random.choice(response_list)}


```

```python
class dupPromptTemplate:
    
    def __init__(self,template,input_variables):
        self.template=template
        self.input_variables=input_variables
    def format(self,input_dict):
        return self.template.format(**input_dict)
    
```

```python
pt=dupPromptTemplate(
    template="Write a {length} essay about {topic}",
    input_variables={"length","topic"}
)

prompt=pt.format({"length":"short","topic":"AI agents"})
```

```python
llm=dupLLM()
```

```python
llm.predict(prompt)
```

```python
class dupLLMChain:
    def __init__(self,llm,prompt):
        self.llm=llm
        self.prompt=prompt
    def run(self,input_dict):
        final_prompt=self.prompt.format(input_dict)
        result=self.llm.predict(final_prompt)
        return result["response"]
```

```python
template=dupPromptTemplate(
    template="Write a {length} essay about {topic}",
    input_variables=["length","topic"]
)

```

```python
llm=dupLLM()
```

```python
chain=dupLLMChain(llm,template)
```
